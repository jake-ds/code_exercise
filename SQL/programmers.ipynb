#### 중복제거하기

```python
"""
select count (distinct cut)\
  from data
"""

pdsql.sqldf("select count (distinct cut)\
  from data")
```

```python
"""
SELECT animal_type, count(*)\
  from animal_ins\
 group by animal_type
"""

pdsql.sqldf("SELECT cut, count(*)\
  from data\
 group by cut")
```

```python
"""
select A.*\
   from (select name, count(*) as COUNT\
           from animal_ins\
group by name) AS A\
where A.count>1
"""

pdsql.sqldf("select A.*\
               from (select cut, count(*) as COUNT\
                       from data group by cut) AS A\
              where A.count>1")
```

```python
"""
select name, count(*)/
  from animal_ins/
 group by name/
having count(name) >1/
 order by name
"""

pdsql.sqldf("select cut, count(*)\
              from data\
             group by cut\
            having count(cut) >10000\
             order by cut")
```
